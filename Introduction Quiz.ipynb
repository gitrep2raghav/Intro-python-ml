```python
import pandas as pd
import numpy as np
import seaborn as sn
import matplotlib.pyplot as plt
```

```python
import pandas as pd
csv_df = pd.read_csv( 'FoodBalanceSheets_E_Africa_NOFLAG.csv', encoding="ISO-8859-1")
csv_df.to_csv( 'FoodBalanceSheets_E_Africa_NOFLAG.csv' , index= False )
```

```python
#What is the total sum of Animal Fat produced in 2014 and 2017 respectively?
pd.set_option('display.max_rows', 1000)
booleans = []
for item in csv_df.Item:
    if item=='Animal fats':
        booleans.append(True)
    else:
        booleans.append(False)
is_Anima = pd.Series(booleans)
csv_df[is_Anima].groupby(['Item'])['Y2014','Y2017'].sum()   
```

```python
#What is the mean and standard deviation across the whole dataset for the year 2015 to 3 decimal places?
mean1 = csv_df['Y2015'].mean()
std1 = csv_df['Y2015'].std()
print("Mean is " +str(format(mean1,"0.3f"))+"\nStd Deviation is "+str(format(std1,"0.3f")))
```

```python
#What is the total number and percentage of missing data in 2016 to 2 decimal places?
null1 = csv_df['Y2016'].isna().sum()
total_rows = csv_df['Y2016'].count()
print(null1)
print(format(null1/total_rows*100,'0.2f'))
```

```python
#Which year had the highest correlation with ‘Element Code’? 
corrmatrix=csv_df.corr()
sn.heatmap(corrmatrix,annot=True)
print(corrmatrix)
plt.show()
```

```python
#What year has the highest sum of Import Quantity?
booleans = []
for item in csv_df.Element:
    if item=='Import Quantity':
        booleans.append(True)
    else:
        booleans.append(False)
highest_qty = pd.Series(booleans)
csv_df[highest_qty].groupby(['Element'])['Y2014','Y2015','Y2016','Y2017'].sum()
```

```python
#What is the total number of the sum of Production in 2014?
csv_df.groupby(['Element'])['Y2014','Y2015','Y2016','Y2017'].sum()
```

```python
#Which of these elements had the highest sum in 2018?
df=csv_df.groupby(['Element'])['Y2018'].sum()
df1 = pd.DataFrame(df,columns=['Elements','Values'])
df1.sort_values('Values',ascending=False)
```
